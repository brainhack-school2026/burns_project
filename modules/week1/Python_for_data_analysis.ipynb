```python
# imports

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
iris = load_iris()
```

```python
iris.keys()
```

```python
iris_type = type(iris['data'])
iris_shape = np.shape(iris['data'])

print(f"Data shape: {iris_shape}, Data type: {iris_type}")
```

```python
data = iris['data']
feature_names = iris['feature_names']
```

```python
new_df = pd.DataFrame(data=data, columns=feature_names)
```

```python
new_df.describe()
```

```python
new_df_50 = new_df.loc[0:50]
```

```python
mean = new_df['sepal length (cm)'].mean()
std = new_df['sepal length (cm)'].std()

extreme_count = 0
for length in new_df['sepal length (cm)']:
    if (length-mean)/std > 3.9:
        print(length)
        extreme_count += 1

print(extreme_count)

        
```

```python
def find_extreme_values(column):
    mean = column.mean()
    std = column.std()
    
    extreme_count = 0
    for length in column:
        if abs((length-mean)/std) > 3.9:
            print(length)
            extreme_count += 1
    
    print(extreme_count)
```

```python
for col in new_df.columns:
    find_extreme_values(new_df[col])
```

```python
fig, ax = plt.subplots(figsize=(15,5))
plt.boxplot(x=new_df, tick_labels=feature_names)
ax.set_ylabel('distance (cm)')
ax.set_title('Box plot distribution')
plt.show()
```

```python
new_df.to_csv('iris_df.csv')
fig.savefig('iris_boxplot.png')
```
